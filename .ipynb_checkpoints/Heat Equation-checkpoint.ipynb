# Solving the Heat Equation: $\Phi$<sub>'tt</sub> = $\Phi$<sub>'xx</sub> + $\beta$$\Phi$

## Introduction:

The following code finds the solution to the heat equation with a potential term $\beta\Phi$ and with no incoming boundary values. It separates the second order PDE into first order PDEs and uses the forward-time-centered-space (FTCS) scheme to compute the solution. It the computes the convergence of the FTCS scheme. The solution is then plotted for different values of $\beta$.

The final two sections introduce the solution to $\Phi$<sub>'tt</sub> = $\Phi$<sub>'xx</sub> + sin($\Phi$) and  the Runge-Kutta RK3 method for solving the original pde. RK3 should be unconditionally stable, but there is an error in the code that makes the solution found using this code unstable.


## The FTCS Scheme:

```julia
using Plots
```

First, create a function for the initial conditions when the function is zero at the boundary. The function take the x position as input and outputs the initial conditions at that point. 

```julia
# Initial conditions
# phi,tt = phi,xx
# g = phi,t
# f = phi,x

# phi,t = g
# f,t = g,x
# g,t = f,x
function init(x::Float64)
    phi_initial = exp(-((x-5)^2)) - exp(-(5)^2)
    f_initial = exp(-((x-5)^2)) - exp(-(5)^2)
    g_initial = -2*(x-5)*exp(-((x-5)^2))
    return [phi_initial,g_initial,f_initial]
end
```

Next create a function for the FTCS scheme, the function take the initial values of $\Phi$, f and g (U0, f0, and g0) as well as the timestep (dt) the space step (dx) and the value of beta (B), and outputs the values of $\Phi$, f and g at the next timestep.

```julia
# Solid boundary contitions

function update(U0,f0,g0,dt::Float64,dx::Float64,B::Float64)
    
    n = length(U0)
    
    # Initialize arrays for next timestep
    
    UN = zeros(n,1)
    fN = zeros(n,1)
    gN = zeros(n,1)
    
    # Find the solution at the left boundary
    
    UN[1] = 0
    fN[1] = 0
    gN[1] = g0[1]+ dt*(f0[2]-f0[1])/dx
    
    for i in 2:n-1
        
        # Find solution for middle using FTCS

        fN[i] = f0[i] + dt*(g0[i+1]-g0[i-1])/(2*dx) - dt*B^2*U0[i]
        gN[i] = g0[i] + dt*(f0[i+1]-f0[i-1])/(2*dx)
        UN[i] = U0[i] + dt*f0[i]

    end
    
    # Find solution at right boundary
    
    UN[n] = 0
    fN[n] = 0
    gN[n] = g0[n]+ dt*(f0[n]-f0[n-1])/dx
    
    return UN,fN,gN
    
end
```

Now initialize the function and the grid spacings.

```julia
# Length of x grid:

xstart = 0
xend = 10

#Number of grid points and value of lambda:

N = 100
lambda = 0.1

# Determine grid spacings:

delta_x = (xend-xstart)/N
delta_t = lambda*delta_x

beta = 1.0;
   
# Initialize arrays for phi, f and g:

phi0 = zeros(N+1,1)
f0 = zeros(N+1,1)
g0 = zeros(N+1,1)

# Fill the arrays

for i in 1:N+1    
    phi0[i,1],g0[i,1],f0[i,1] = init(delta_x*(i-1))
end

# Fill an array with the x values

x = zeros(N+1,1)
for i in 1:N+1
    x[i] = delta_x*(i-1)
end
```

Now use the FTCS scheme to solve the equation and plot the solution:

```julia
j = 0
p = Array{Any,2}(undef,100,2)

for i in 1:6000

    # Update the function
    
    phiN,fN,gN = update(phi0,f0,g0,delta_t,delta_x,beta)
    for j in 1:N+1
        phi0[j] = phiN[j]
        f0[j] = fN[j]
        g0[j] = gN[j]
    end
    
    # Save plots to make into a gif
    
    if i%60 == 0
        j = j+1
        p[j] = plot(x,phi0,ylims=[-1,1],xlabel="x",ylabel="Phi",title="Solution when beta=1")
    end
end
```

```julia
# Make the plots into a gif. If @animate doesn't work, type run(`ffmpeg -version`) and check that ffmpeg is installed

anim = @animate for i=1:100
    plot(p[i])
end

gif(anim, "C:\\Users\\court\\AppData\\Local\\Temp\\test.gif", fps = 5)
```

## Convergence

The following is a function to compute the convergence of a given scheme, udate, with a given lambda, and a given number of timesteps and spacesteps, T and N respectively. The value of $\beta$ is always one for this calculation. The function outputs the convergence order, $q$ where $\frac{L_1}{L_2} = 2^q$ and $L_1$ and $L_2$ are the averages of the squared differences between solutions with timesteps of $\tau$ and $\frac{\tau}{2}$ and between solutions with timesteps of $\frac{\tau}{2}$ and $\frac{\tau}{4}$ respectively.

```julia
function convergence(udate,lambda,T,N,beta)
    
    # Length of x grid
    
    xstart = 0
    xend = 10

    # Width of space grid and time grid
    
    delta_x = (xend-xstart)/N

    delta_t1 = lambda*delta_x
    delta_t2 = delta_t1/2
    delta_t3 = delta_t2/2
   
    # Initialize the functions for each timestep width
    
    phi0 = zeros(N+1,3)
    f0 = zeros(N+1,3)
    g0 = zeros(N+1,3)

    for i in 1:N+1
            
        phi0[i,1],g0[i,1],f0[i,1] = init(delta_x*(i-1))
        phi0[i,2],g0[i,2],f0[i,2] = init(delta_x*(i-1))
        phi0[i,3],g0[i,3],f0[i,3] = init(delta_x*(i-1))
    
    end

    # Update for each timestep width
    
    for i in 1:T

        phiN1,fN1,gN1 = udate(phi0[1:N+1,1],f0[1:N+1,1],g0[1:N+1,1],delta_t1,delta_x,beta)

        for j in 1:N+1
            phi0[j,1] = phiN1[j]
            f0[j,1] = fN1[j]
            g0[j,1] = gN1[j]
        end
        
    end

    for i in 1:2*T

        phiN2,fN2,gN2 = udate(phi0[1:N+1,2],f0[1:N+1,2],g0[1:N+1,2],delta_t2,delta_x,beta)

        for j in 1:N+1
            phi0[j,2] = phiN2[j]
            f0[j,2] = fN2[j]
            g0[j,2] = gN2[j]
        end
        
    end

    for i in 1:4*T

        phiN3,fN3,gN3 = udate(phi0[1:N+1,3],f0[1:N+1,3],g0[1:N+1,3],delta_t3,delta_x,beta)

        for j in 1:N+1
            phi0[j,3] = phiN3[j]
            f0[j,3] = fN3[j]
            g0[j,3] = gN3[j]
        end
        
    end

    # Finding the average squared differences
    
    L1 = sqrt((sum(phi0[:,1]-phi0[:,2]))^2/N)
    L2 = sqrt((sum(phi0[:,2]-phi0[:,3]))^2/N)

    # Computing C
    
    converge = L1/L2
    
    # Computing q
    
    conv = log2(converge)
    
    return conv
    
end
    
```

Now the convergence can be computed for varying values of $\lambda$, beta, and different numbers of timesteps. First we'll vary lambda:

```julia
lambda = zeros(50,1)
cnvg = zeros(50,1)

for i in 1:50
    lambda[i] = 0.005*i
    cnvg[i] = convergence(update,lambda[i],10000,100,1.0)
end
```

```julia
plot(lambda,cnvg,yaxis=:log,xlabel="lambda",ylabel="convergence factor",title="Log plot of lambda vs. convergence factor \n beta=1 and 10 000 timesteps", legend=false)
```

So when we take 10 000 timesteps and vary $\lambda$, the solution explodes around $\lambda=0.075$. Now let's vary the timesteps:

```julia
lambda = zeros(100,1)
T = zeros(100,1)
cnvg = zeros(100,1)

for i in 1:100
    T[i] = 1000*i
    cnvg[i] = convergence(update,0.05,T[i],100,1.0)
end
```

```julia
plot(T,cnvg,xlabel="number of timesteps",ylabel="convergence factor",title="Plot of number of timesteps vs. convergence factor \n lambda = 0.05, beta=1",legend=false)
```

So for $\lambda=0.05$ the solution stays reasonably stable up to around 25 000  timesteps. Looking at the document entitled "Stability Analysis.pdf" in the github repository for this project, it can be seen that the the FTCS scheme is unconditionally unstable for all $\lambda$.

```julia
beta = zeros(100,1)
cnvg = zeros(100,1)

for i in 1:100
    beta[i] = 0.1*i
    cnvg[i] = convergence(update,0.05,10000,100,beta[i])
end

plot(beta,cnvg,xlabel="beta",ylabel="convergence factor",title="Plot of beta vs. convergence factor \n lambda=0.05, 10 000 timesteps",legend=false)
```

So the solution slowly becomes less and less convergent as beta increases. The effect is not as dramatic as when the number of timesteps or the value of $\lambda$ is changed, but it is still noticable.

## Solution to $\Phi$<sub>'tt</sub> = $\Phi$<sub>'xx</sub> + sin($\Phi$)

The following code shows the solution to the above PDE which is a variation on the original PDE.

```julia
# Solid boundary contitions

function update2(U0,f0,g0,dt::Float64,dx::Float64,B::Float64)
    
    n = length(U0)
    
    # Initialize arrays for next timestep
    
    UN = zeros(n,1)
    fN = zeros(n,1)
    gN = zeros(n,1)
    
    # Find the solution at the left boundary
    
    UN[1] = 0
    fN[1] = 0
    gN[1] = g0[1]+ dt*(f0[2]-f0[1])/dx
    
    for i in 2:n-1
        
        # Find solution for middle using FTCS

        fN[i] = f0[i] + dt*(g0[i+1]-g0[i-1])/(2*dx) - dt*sin(U0[i])
        gN[i] = g0[i] + dt*(f0[i+1]-f0[i-1])/(2*dx)
        UN[i] = U0[i] + dt*f0[i]

    end
    
    # Find solution at right boundary
    
    UN[n] = 0
    fN[n] = 0
    gN[n] = g0[n]+ dt*(f0[n]-f0[n-1])/dx
    
    return UN,fN,gN
    
end
```

```julia
# Length of x grid

xstart = 0
xend = 10

# Number of space steps and value of lambda

N = 100
lambda = 0.1

# Grid sizes

delta_x = (xend-xstart)/N

delta_t = lambda*delta_x
   
#initialize arrays and input values

phi0 = zeros(N+1,1)
f0 = zeros(N+1,1)
g0 = zeros(N+1,1)

for i in 1:N+1    
    phi0[i,1],g0[i,1],f0[i,1] = init(delta_x*(i-1))
end

x = zeros(N+1,1)
for i in 1:N+1
    x[i] = delta_x*(i-1)
end
```

```julia
# Solve the PDE and save plots to turn into a gif

j = 0
p = Array{Any,2}(undef,100,2)

for i in 1:6000

    phiN,fN,gN = update2(phi0,f0,g0,delta_t,delta_x,beta)
    for j in 1:N+1
        phi0[j] = phiN[j]
        f0[j] = fN[j]
        g0[j] = gN[j]
    end
    
    if i%60 == 0
        j = j+1
        p[j] = plot(x,phi0,ylims=[-1,1],xlabel="x",ylabel="Phi",title="Solution of PDE with sin term",legend=false)
    end
end
```

## Runge-Kutta

The following code solves the original PDE using the RK3 scheme. Unfortunately there is an error somewhere in the code that makes the solution explode at the boundary.

```julia
# Solid boundary contitions

function updateRK2(U0,f0,g0,dt::Float64,dx::Float64,B::Float64)
    
    n = size(U0)[1]
    
    # Initialize arrays for the final solutions, the different k's and the intermediate solutions
    
    UN = zeros(n,1)
    fN = zeros(n,1)
    gN = zeros(n,1)
       
    k1_f = zeros(n,1)
    k2_f = zeros(n,1)
    k3_f = zeros(n,1)
 
    f1 = zeros(n,1) # These are intermediate solutions
    f2 = zeros(n,1)
    
    k1_g = zeros(n,1)
    k2_g = zeros(n,1)
    k3_g = zeros(n,1)
    
    g1 = zeros(n,1) # These are intermediate solutions
    g2 = zeros(n,1)
    
    k1_U = zeros(n,1)
    k2_U = zeros(n,1)
    k3_U = zeros(n,1)
    
    U1 = zeros(n,1)
    U2 = zeros(n,1)
    
    UN[1] = 0
    fN[1] = 0
    gN[1] = g0[1]+ dt*(f0[2]-f0[1])/dx
    
    # Calculate the k1's and the first intermediate solutions
    
    for i in 2:n-1
        
        k1_f[i] = (g0[i+1] - g0[i-1])/(2*dx)
        k1_g[i] = (f0[i+1] - f0[i-1])/(2*dx) - (B^2)*U0[i]
        k1_U[i] = f0[i]
        
        f1[i] = f0[i] + dt*k1_f[i]
        g1[i] = g0[i] + dt*k1_g[i]
        U1[i] = U0[i] + dt*k1_U[i]
        
    end
    
    # Calculate the k2's and the second intermediate solutions
    
    for i in 4:n-3
        
        k2_f[i] = ((g0[i+1]-g0[i-1])+(dt/(4*dx))*(g1[i+2]-2*g1[i]+g1[i-2]))/(2*dx)
        k2_g[i] = ((f0[i+1]-f0[i-1])+(dt/(4*dx))+(f1[i+2]-2*f1[i]+f1[i-2]))/(2*dx)+(dt*(B^2)/(4*dx))*(U1[i-1]-U1[i+1])
        k2_U[i] = f0[i] + (dt/2)f1[i]
        
        f2[i] = f0[i] + dt*k2_f[i]
        g2[i] = g0[i] + dt*k2_g[i]
        U2[i] = U0[i] + dt*k2_U[i]
        
    end
    
    # Calculate the k3's and the final solutions
    
    for i in 7:n-6
        k3_f[i] = ((g0[i+1]-g0[i-1])/(2*dx)) + ((dt/(4*dx^2))*(g2[i+2]-2*g2[i]+g2[i-2])) + ((dt^2)/(8*dx^3))*(g2[i+3]-3*g2[i+1]+3*g2[i-1]-g2[i-3])
        fN[i] = f0[i] + (dt/6)*(k1_f[i] + (4*k2_f[i]) + k3_f[i])
        
        k3_g[i] = ((f0[i+1]-f0[i-1])/(2*dx)) + ((dt/(4*dx^2))*(f2[i+2]-2*f2[i]+f2[i-2])) + ((dt^2)/(8*dx^3))*(f2[i+3]-3*f2[i+1]+3*f2[i-1]-f2[i-3])+ ((dt^2)*(B^2)/(4*dx^2))*(2U2[i]-U2[i+2]-U2[i-2])+((dt^2)*(B^2)/(2*dx))*(U2[i-1]-U2[i+1])-2*dt*(B^2)U2[i]                                                                        
        gN[i] = g0[i] + (dt/6)*(k1_g[i] + (4*k2_g[i]) + k3_g[i])
        
        k3_U[i] = f0[i] + dt*f2[i]
        UN[i] = U0[i] + (dt/6)*(k1_U[i] + (4*k2_U[i]) + k3_U[i])
    end

    # Near the boundary, it is necessary to calculate the solution using RK2 and FTCS since RK3 requires too many
    # original points for each solution point
    
    for i in 4:6
        
        # Solutions using RK2
        
        fN[i] = f2[i]
        gN[i] = g2[i]
        UN[i] = U2[i]
    end
    
    for i in n-5:n-3
        
        # Solutions using RK2
        
        fN[i] = f2[i]
        gN[i] = g2[i]
        UN[i] = U2[i]
    end
    
    for i in 2:3
        
        # Solutions using FTCS
        
        fN[i] = f1[i]
        gN[i] = g1[i]
        UN[i] = U1[i]
    end
    
    for i in n-2:n-1
        
        # Solutions using FTCS
        
        fN[i] = f1[i]
        gN[i] = g1[i]
        UN[i] = U1[i]
    end
    
    # Solution at boundary
    
    UN[n] = 0
    fN[n] = 0
    gN[n] = g0[n]+ dt*(f0[n]-f0[n-1])/dx
    
    return UN,fN,gN
    
end
```

```julia
phi0 = zeros(N+1,1)
f0 = zeros(N+1,1)
g0 = zeros(N+1,1)

for i in 1:N+1
            
    phi0[i,1],g0[i,1],f0[i,1] = init(delta_x*(i-1))
            
end

x = zeros(N+1,1)
for i in 1:N+1
    x[i] = delta_x*(i-1)
end
```

```julia
j = 0
p = Array{Any,2}(undef,10,2)

for i in 1:200

    phiN,fN,gN = updateRK2(phi0,f0,g0,delta_t,delta_x,beta)
    for j in 1:N+1
        phi0[j] = phiN[j]
        f0[j] = fN[j]
        g0[j] = gN[j]
    end
    
    if i%20 == 0
        j = j+1
        p[j] = plot(x,phi0,ylim=[-1.5,1.5],xlabel="x",ylabel="Phi",title="Solution using RK3 scheme",legend=false)
    end
end
```

## Conclusion

The FTCS scheme is unconditionally unstable, as shown by the convergence plots. This was expected from the stability caluclations done in the file "Stability Analysis.pdf". The
