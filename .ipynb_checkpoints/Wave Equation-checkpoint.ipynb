# Solve $\Phi$<sub>'tt</sub>=$\Phi$<sub>'xx</sub>+$\beta$$\Phi$

#### Solve on a periodic grid then solve on x is an element of [0,L] and impose no incoming boundary values at both sides (U(0)=U(L)=0). Then show convergence of the obtained solution.

```julia
using Plots
```

```julia
xstart = 0
xend = 10

N = 100

delta_x = (xend-xstart)/N

lambda = 0.1

delta_t = lambda*delta_x

beta = 1.0;
```

```julia
# Initial conditions
# phi,tt = phi,xx
# g = phi,t
# f = phi,x

# phi,t = g
# f,t = g,x
# g,t = f,x <- find this by plugging into the original equation 
function init(x::Float64)
    phi_initial = exp(-((x-5)^2)) - exp(-(5)^2)
    f_initial = exp(-((x-5)^2)) - exp(-(5)^2)
    g_initial = -2*(x-5)*exp(-((x-5)^2))
    return [phi_initial,g_initial,f_initial]
end
```

```julia
phi0 = zeros(N+1,1)
f0 = zeros(N+1,1)
g0 = zeros(N+1,1)

for i in 1:N+1
    phi0[i],g0[i],f0[i] = init(delta_x*(i-1))
end

x = zeros(N+1,1)
for i in 1:N+1
    x[i] = delta_x*(i-1)
end
```

```julia
p1 = plot(x,phi0)
p2 = plot(x,f0)
p3 = plot(x,g0)
plot(p1,p2,p3,layout=(1,3),legend=false)
```

```julia
# Solid boundary contitions

function update(U0::Array{Float64,2},f0::Array{Float64,2},g0::Array{Float64,2},dt::Float64,dx::Float64,B::Float64)
    
    UN = zeros(length(U0),1)
    fN = zeros(length(U0),1)
    gN = zeros(length(U0),1)
    
    UN[1] = 0
    fN[1] = 0
    gN[1] = g0[1]+ dt*(f0[2]-f0[1])/dx
    
    for i in 2:100
        
        # Forward Time Centered Space

        fN[i] = f0[i] + dt*(g0[i+1]-g0[i-1])/(2*dx) - dt*B^2*U0[i]
        gN[i] = g0[i] + dt*(f0[i+1]-f0[i-1])/(2*dx)
        UN[i] = U0[i] + dt*f0[i]

    end
    
    UN[101] = 0
    fN[101] = 0
    gN[101] = g0[101]+ dt*(f0[101]-f0[100])/dx
    
    return UN,fN,gN
    
end
```

```julia
j = 0
p = Array{Any,2}(undef,10,2)

for i in 1:1000

    phiN,fN,gN = update(phi0,f0,g0,delta_t,delta_x,beta)
    for j in 1:101
        phi0[j] = phiN[j]
        f0[j] = fN[j]
        g0[j] = gN[j]
    end
    
    if i%100 == 0
        j = j+1
        p[j] = plot(x,phi0)
    end
end
```

```julia
plot(p[1],p[2],p[3],p[4],p[5],p[6],p[7],p[8],p[9],p[10],layout=(5,2),legend=false)
```

```julia
phi0 = zeros(N+1,1)
f0 = zeros(N+1,1)
g0 = zeros(N+1,1)

for i in 1:N+1
    phi0[i],g0[i],f0[i] = init(delta_x*(i-1))
end

x = zeros(N+1,1)
for i in 1:N+1
    x[i] = delta_x*(i-1)
end
```

```julia
# Periodic boundary contitions

function update(U0::Array{Float64,2},f0::Array{Float64,2},g0::Array{Float64,2},dt::Float64,dx::Float64)
    
    UN = zeros(length(U0),1)
    fN = zeros(length(U0),1)
    gN = zeros(length(U0),1)
    
    UN[1] = 0
    fN[1] = 0
    gN[1] = 0
    
    for i in 2:100
        
        # Centered

        fN[i] = f0[i] + dt*(g0[i+1]-g0[i-1])/(2dx)
        gN[i] = g0[i] + dt*(f0[i+1]-f0[i-1])/(2dx)
        UN[i] = U0[i] + dt*(g0[i+1]-g0[i-1])/(2dx)

    end
    
    UN[101] = 0
    fN[101] = 0
    gN[101] = 0
    
    return UN,fN,gN
    
end
```
