# Solving the Heat Equation: $\Phi$<sub>'tt</sub> = $\Phi$<sub>'xx</sub> + $\beta$$\Phi$

## Introduction:

The following code finds the solution to the heat equation with no incoming boundary values. It uses the forward-time-centered-space scheme and computes the convergence of the method. The solution is then plotted for different values of $\beta$.

The final section introduces the Runge-Kutta RK3 method for solving the pde. RK3 should be unconditionally stable, but there is an error in the code that makes the solution unstable.


## The FTCS Scheme:

```julia
using Plots
```

First, create a function to create the initial conditions when the function is zero at the boundary.

```julia
# Initial conditions
# phi,tt = phi,xx
# g = phi,t
# f = phi,x

# phi,t = g
# f,t = g,x
# g,t = f,x <- find this by plugging into the original equation 
function init(x::Float64)
    phi_initial = exp(-((x-5)^2)) - exp(-(5)^2)
    f_initial = exp(-((x-5)^2)) - exp(-(5)^2)
    g_initial = -2*(x-5)*exp(-((x-5)^2))
    return [phi_initial,g_initial,f_initial]
end
```

Next create a function for the FTCS scheme:

```julia
# Solid boundary contitions

function update(U0,f0,g0,dt::Float64,dx::Float64,B::Float64)
    
    n = length(U0)
    
    UN = zeros(n,1)
    fN = zeros(n,1)
    gN = zeros(n,1)
    
    UN[1] = 0
    fN[1] = 0
    gN[1] = g0[1]+ dt*(f0[2]-f0[1])/dx
    
    for i in 2:n-1
        
        # Forward Time Centered Space

        fN[i] = f0[i] + dt*(g0[i+1]-g0[i-1])/(2*dx) - dt*B^2*U0[i]
        gN[i] = g0[i] + dt*(f0[i+1]-f0[i-1])/(2*dx)
        UN[i] = U0[i] + dt*f0[i]

    end
    
    UN[n] = 0
    fN[n] = 0
    gN[n] = g0[n]+ dt*(f0[n]-f0[n-1])/dx
    
    return UN,fN,gN
    
end
```

Now initialize the function and the grid spacings.

```julia
xstart = 0
xend = 10

N = 100
lambda = 0.1

delta_x = (xend-xstart)/N

delta_t = lambda*delta_x

beta = 1.0;
   
phi0 = zeros(N+1,1)
f0 = zeros(N+1,1)
g0 = zeros(N+1,1)

for i in 1:N+1    
    phi0[i,1],g0[i,1],f0[i,1] = init(delta_x*(i-1))
end

x = zeros(N+1,1)
for i in 1:N+1
    x[i] = delta_x*(i-1)
end
```

Now use the FTCS scheme to solve the equation and plot the solution:

```julia
j = 0
p = Array{Any,2}(undef,100,2)

for i in 1:6000

    phiN,fN,gN = update(phi0,f0,g0,delta_t,delta_x,beta)
    for j in 1:N+1
        phi0[j] = phiN[j]
        f0[j] = fN[j]
        g0[j] = gN[j]
    end
    
    if i%60 == 0
        j = j+1
        p[j] = plot(x,phi0,ylims=[-1,1])
    end
end
```

## Convergence

```julia
function convergence(udate,lambda,T,N)
    
    xstart = 0
    xend = 10

    delta_x = (xend-xstart)/N

    delta_t1 = lambda*delta_x
    delta_t2 = delta_t1/2
    delta_t3 = delta_t2/2

    beta = 1.0;
   
    phi0 = zeros(N+1,3)
    f0 = zeros(N+1,3)
    g0 = zeros(N+1,3)

    for i in 1:N+1
            
        phi0[i,1],g0[i,1],f0[i,1] = init(delta_x*(i-1))
        phi0[i,2],g0[i,2],f0[i,2] = init(delta_x*(i-1))
        phi0[i,3],g0[i,3],f0[i,3] = init(delta_x*(i-1))
    
    end

    for i in 1:T

        phiN1,fN1,gN1 = udate(phi0[1:N+1,1],f0[1:N+1,1],g0[1:N+1,1],delta_t1,delta_x,beta)

        for j in 1:N+1
            phi0[j,1] = phiN1[j]
            f0[j,1] = fN1[j]
            g0[j,1] = gN1[j]
        end
        
    end

    for i in 1:2*T

        phiN2,fN2,gN2 = udate(phi0[1:N+1,2],f0[1:N+1,2],g0[1:N+1,2],delta_t2,delta_x,beta)

        for j in 1:N+1
            phi0[j,2] = phiN2[j]
            f0[j,2] = fN2[j]
            g0[j,2] = gN2[j]
        end
        
    end

    for i in 1:4*T

        phiN3,fN3,gN3 = udate(phi0[1:N+1,3],f0[1:N+1,3],g0[1:N+1,3],delta_t3,delta_x,beta)

        for j in 1:N+1
            phi0[j,3] = phiN3[j]
            f0[j,3] = fN3[j]
            g0[j,3] = gN3[j]
        end
        
    end

    L1 = sqrt((sum(phi0[:,1]-phi0[:,2]))^2/N)
    L2 = sqrt((sum(phi0[:,2]-phi0[:,3]))^2/N)

    converge = L1/L2
    
    conv = log2(converge)
    
    return conv
    
end
    
```

```julia
lambda = zeros(50,1)
cnvg = zeros(50,1)

for i in 1:50
    lambda[i] = 0.005*i
    cnvg[i] = convergence(update,lambda[i],10000,100)
end
```

```julia
plot(lambda,cnvg,yaxis=:log,xlabel="lambda",ylabel="convergence factor",title="Log plot of lambda vs. convergence factor", legend=false)
```

```julia
lambda = zeros(100,1)
T = zeros(100,1)
cnvg = zeros(100,1)

for i in 1:100
    T[i] = 100*i
    cnvg[i] = convergence(update,0.05,T[i],100)
end
```

```julia
plot(T,cnvg,xlabel="number of timesteps",ylabel="convergence factor",title="Plot of number of timesteps vs. convergence factor",legend=false)
```

## Solution to $\Phi$<sub>'tt</sub> = $\Phi$<sub>'xx</sub> + sin($\Phi$)

```julia
# Solid boundary contitions

function update2(U0,f0,g0,dt::Float64,dx::Float64,B::Float64)
    
    n = length(U0)
    
    UN = zeros(n,1)
    fN = zeros(n,1)
    gN = zeros(n,1)
    
    UN[1] = 0
    fN[1] = 0
    gN[1] = g0[1]+ dt*(f0[2]-f0[1])/dx
    
    for i in 2:n-1
        
        # Forward Time Centered Space

        fN[i] = f0[i] + dt*(g0[i+1]-g0[i-1])/(2*dx) - dt*sin(U0[i])
        gN[i] = g0[i] + dt*(f0[i+1]-f0[i-1])/(2*dx)
        UN[i] = U0[i] + dt*f0[i]

    end
    
    UN[n] = 0
    fN[n] = 0
    gN[n] = g0[n]+ dt*(f0[n]-f0[n-1])/dx
    
    return UN,fN,gN
    
end
```

```julia
xstart = 0
xend = 10

N = 100
lambda = 0.1

delta_x = (xend-xstart)/N

delta_t = lambda*delta_x

beta = 1.0;
   
phi0 = zeros(N+1,1)
f0 = zeros(N+1,1)
g0 = zeros(N+1,1)

for i in 1:N+1    
    phi0[i,1],g0[i,1],f0[i,1] = init(delta_x*(i-1))
end

x = zeros(N+1,1)
for i in 1:N+1
    x[i] = delta_x*(i-1)
end
```

## Runge-Kutta

```julia
# Solid boundary contitions

function updateRK2(U0,f0,g0,dt::Float64,dx::Float64,B::Float64)
    
    n = size(U0)[1]
    
    UN = zeros(n,1)
    fN = zeros(n,1)
    gN = zeros(n,1)
    
    UN[1] = 0
    fN[1] = 0
    gN[1] = g0[1]+ dt*(f0[2]-f0[1])/dx
    
    k1_f = zeros(n,1)
    k2_f = zeros(n,1)
    k3_f = zeros(n,1)
 
    f1 = zeros(n,1)
    f2 = zeros(n,1)
    
    k1_g = zeros(n,1)
    k2_g = zeros(n,1)
    k3_g = zeros(n,1)
    
    g1 = zeros(n,1)
    g2 = zeros(n,1)
    
    k1_U = zeros(n,1)
    k2_U = zeros(n,1)
    k3_U = zeros(n,1)
    
    U1 = zeros(n,1)
    U2 = zeros(n,1)
    
    for i in 2:n-1
        
        k1_f[i] = (g0[i+1] - g0[i-1])/(2*dx)
        k1_g[i] = (f0[i+1] - f0[i-1])/(2*dx) - (B^2)*U0[i]
        k1_U[i] = f0[i]
        
        f1[i] = f0[i] + dt*k1_f[i]
        g1[i] = g0[i] + dt*k1_g[i]
        U1[i] = U0[i] + dt*k1_U[i]
        
    end
    
    for i in 4:n-3
        
        k2_f[i] = ((g0[i+1]-g0[i-1])+(dt/(4*dx))*(g1[i+2]-2*g1[i]+g1[i-2]))/(2*dx)
        k2_g[i] = ((f0[i+1]-f0[i-1])+(dt/(4*dx))+(f1[i+2]-2*f1[i]+f1[i-2]))/(2*dx)+(dt*(B^2)/(4*dx))*(U1[i-1]-U1[i+1])
        k2_U[i] = f0[i] + (dt/2)f1[i]
        
        f2[i] = f0[i] + dt*k2_f[i]
        g2[i] = g0[i] + dt*k2_g[i]
        U2[i] = U0[i] + dt*k2_U[i]
        
    end
    
    for i in 7:n-6
        k3_f[i] = ((g0[i+1]-g0[i-1])/(2*dx)) + ((dt/(4*dx^2))*(g2[i+2]-2*g2[i]+g2[i-2])) + ((dt^2)/(8*dx^3))*(g2[i+3]-3*g2[i+1]+3*g2[i-1]-g2[i-3])
        fN[i] = f0[i] + (dt/6)*(k1_f[i] + (4*k2_f[i]) + k3_f[i])
        
        k3_g[i] = ((f0[i+1]-f0[i-1])/(2*dx)) + ((dt/(4*dx^2))*(f2[i+2]-2*f2[i]+f2[i-2])) + ((dt^2)/(8*dx^3))*(f2[i+3]-3*f2[i+1]+3*f2[i-1]-f2[i-3])+ ((dt^2)*(B^2)/(4*dx^2))*(2U2[i]-U2[i+2]-U2[i-2])+((dt^2)*(B^2)/(2*dx))*(U2[i-1]-U2[i+1])-2*dt*(B^2)U2[i]                                                                        
        gN[i] = g0[i] + (dt/6)*(k1_g[i] + (4*k2_g[i]) + k3_g[i])
        
        k3_U[i] = f0[i] + dt*f2[i]
        UN[i] = U0[i] + (dt/6)*(k1_U[i] + (4*k2_U[i]) + k3_U[i])
    end

    for i in 4:6
        fN[i] = f2[i]
        gN[i] = g2[i]
        UN[i] = U2[i]
    end
    
    for i in n-5:n-3
        fN[i] = f2[i]
        gN[i] = g2[i]
        UN[i] = U2[i]
    end
    
    for i in 2:3
        fN[i] = f1[i]
        gN[i] = g1[i]
        UN[i] = U1[i]
    end
    
    for i in n-2:n-1
        fN[i] = f1[i]
        gN[i] = g1[i]
        UN[i] = U1[i]
    end
    
    UN[n] = 0
    fN[n] = 0
    gN[n] = g0[n]+ dt*(f0[n]-f0[n-1])/dx
    
    return UN,fN,gN
    
end
```

```julia
phi0 = zeros(N+1,1)
f0 = zeros(N+1,1)
g0 = zeros(N+1,1)

for i in 1:N+1
            
    phi0[i,1],g0[i,1],f0[i,1] = init(delta_x*(i-1))
            
end

x = zeros(N+1,1)
for i in 1:N+1
    x[i] = delta_x*(i-1)
end
```

```julia
j = 0
p = Array{Any,2}(undef,10,2)

for i in 1:200

    phiN,fN,gN = updateRK2(phi0,f0,g0,delta_t,delta_x,beta)
    for j in 1:N+1
        phi0[j] = phiN[j]
        f0[j] = fN[j]
        g0[j] = gN[j]
    end
    
    if i%20 == 0
        j = j+1
        p[j] = plot(x,phi0,ylim=[-1.5,1.5])
    end
end
```
